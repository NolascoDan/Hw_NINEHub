```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

link="oec.csv"

data= pd.read_csv(link)
pd.set_option('display.max_columns', None)

data.head()
```

```python
disco_m=data.groupby('DiscoveryMethod').agg({'PlanetaryMassJpt':'sum'}).dropna()
disco_m
```

```python
disco_m.plot.bar()
```
