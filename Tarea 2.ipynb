- Cargue los datos de meteorite_landins.csv de la sesión 3
- Cree un nuevo dataframe llamado meteoritos_h4. En este dataframe, utilizando la función str. seleccione todos los meteoritos solamente de categoría H4
- Cree otro dataframe llamado meteoritos_h5. En este dataframe, utilizando la función str. seleccione todos los meteoritos de solamente categoría H5
- Limpie los datos de ambos dataframes, eliminando las entradas donde las masas sean cero, null, NaN, etc.
- Utilizando masks y selecciones, cuente cuántos meteoritos H4 cayeron entre 1960 y 1990.
- Utilizando masks y selecciones, cuente cuántos meteoritos H5 cayeron entre 1960 y 1990.
- Concatene ambos dataframes y grafique la cantidad de meteoritos por clase y por año

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sb


source= "../Sesión 3/meteorite-landings.csv"

meteo=pd.read_csv(source)
meteo.head()
```

```python
meteoritos_h4= meteo[(meteo.recclass.str.len()==2)&(meteo.recclass.str.contains('H4'))]
meteoritos_h5=meteo[(meteo.recclass.str.contains('H5'))&(meteo.recclass.str.len()==2)]

meteoritos_h4
```

```python
meteoritos_h5
```

```python
meteoritos_h4= meteoritos_h4[(meteoritos_h4['mass']!=0)& (pd.notnull(meteoritos_h4['mass'])) &(pd.notna(meteoritos_h4['mass']))]
meteoritos_h5= meteoritos_h5[(meteoritos_h5['mass']!=0)& (pd.notnull(meteoritos_h5['mass'])) &(pd.notna(meteoritos_h5['mass']))]

meteoritos_h4
```

```python
meteoritos_h5
```

```python
year_h4 = meteoritos_h4[(meteoritos_h4['year'] >= 1960) & (meteoritos_h4['year'] <= 1990)]
year_h4
```

```python
year_h5 = meteoritos_h5[(meteoritos_h5['year'] >= 1960) & (meteoritos_h5['year'] <= 1990)]
year_h5
```

```python
meteoritos_t=pd.concat([year_h4,year_h5])
meteoritos_t
```

```python
p1=year_h4.groupby('year')['recclass'].count()
p2=year_h5.groupby('year')['recclass'].count()
```

```python
p1.plot(color='green')
p2.plot(color='magenta')
pt.title('Meteoritos por año')
```
